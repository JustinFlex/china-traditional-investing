# 04 稳健性与建模（索引）

原 `04_robustness_and_modeling.ipynb` 承担了过多功能（controls / 子样本 / 置换 / 样本外 / Phase 2 / resonance），已拆分为多个单一职责 notebook。

建议运行顺序（可选增强）：
- `03_ganzhi_effect_analysis` 之后
- `04a_controls_models` → `04b_subsample_stability_bing` → `04c_permutation_and_dependence` → `04d_oos_walk_forward`
- Phase 2（交互 gate）：`04e_phase2_interaction_gate`
- Resonance（谐波检验）：`04f_resonance_harmonics`
- 最后 `05_report` 汇总生成报告（不联网，只读落盘文件）

## 04 系列 notebooks
- [`04a_controls_models.ipynb`](04a_controls_models.ipynb)：控制变量回归（OLS/Logit + HAC），产出 `controls_{ts_code}_{group_col}_{target}.csv`
- [`04b_subsample_stability_bing.ipynb`](04b_subsample_stability_bing.ipynb)：子样本分段检验 `stem=丙`，产出 `subsample_bing.csv`
- [`04c_permutation_and_dependence.ipynb`](04c_permutation_and_dependence.ipynb)：置换检验 + HAC 敏感性 + block bootstrap，产出 `perm_* / hac_sensitivity_* / block_bootstrap_*`
- [`04d_oos_walk_forward.ipynb`](04d_oos_walk_forward.ipynb)：按年 walk-forward（样本外），产出 `walk_forward_*`
- [`04e_phase2_interaction_gate.ipynb`](04e_phase2_interaction_gate.ipynb)：Phase 2（`day_group × year_element`）交互 gate，产出 `interaction_* / meta_interaction_*`
- [`04f_resonance_harmonics.ipynb`](04f_resonance_harmonics.ipynb)：`jiazi_idx` k=5/6 谐波检验，产出 `resonance_*`

> 提示：如果你只关心最终报告，通常只需要运行：`04a`（controls）+ `04d`（OOS）+（需要时）`04e/04f`，然后跑 `05_report`。
